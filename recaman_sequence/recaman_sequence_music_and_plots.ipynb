# Recaman Sequence

Let us generate the recaman sequence.

```python
def recaman(n):
    """Produce n first elements of the recaman sequence."""
    seq = [0]
    for step in range(1, n):
        last = seq[-1]
        if (last-step > 1) and (last-step not in seq):
            seq.append(last-step)
        else:
            seq.append(last+step)
    return seq
```

```python
recaman(10)
```

And convert it to the notes:

```python
def int_seq_to_notes(seq, duration=abjad.Duration(1, 16), offset=0):
    notes = []
    for x in seq:
        notes.append(abjad.Note(x+offset, duration))
    return abjad.Container(notes)
```

```python
container = int_seq_to_notes(recaman(50), offset=0)
```

```python
abjad.show(container)
```

```python
abjad.play(container)
```

## Plotting

I also want to try to plot the connections between the consequitive numbers as semi circles (fitting for the pattern of integers).

```python
import matplotlib.pyplot as plt
import matplotlib.patches as ptch
%matplotlib notebook
```

```python
def arc_between_numbers(x1, x2, is_above):
    thetas = (0, 180) if is_above else (180, 360)
    center = ((x1+x2)/2, 0)
    d = abs(x1-x2)
    return ptch.Arc(xy=center, width=d, height=d, theta1=thetas[0], theta2=thetas[1])
```

```python
def show_sequence(seq):
    fig = plt.figure()
    ax = fig.add_subplot()
    
    ax.set_xlim(0, max(seq))
    ax.set_ylim(-max(seq)/3.5, max(seq)/3.5)
    
    is_above = True
    for i in range(len(seq)-1):
        x1 = seq[i]
        x2 = seq[i+1]
        arc = arc_between_numbers(x1, x2, is_above)
        is_above = not is_above
        ax.add_patch(arc)

    plt.show()
```

```python
show_sequence(recaman(100))
```

```python
# Natural numbers
show_sequence(list(range(50)))
```
